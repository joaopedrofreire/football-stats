```python
import pandas as pd

from django.shortcuts import render

import http.client
import json

connection = http.client.HTTPConnection('api.football-data.org')
headers = { 'X-Auth-Token': 'f236d854a2394189a0b6adbab1302b70' }
# Get premier league (2021) data
connection.request('GET', 'api.football-data.org/v2/', None, headers )

# Get matches
matches = json.loads(connection.getresponse().read().decode())['matches']
```

```python
matches
```

```python
data=[] 
for match in matches:
    try:            
        data.append({'matchday': match['matchday'],
                     'date': match['utcDate'],
                     'home_team': match['homeTeam']['name'],
                     'home_team_score': int(match['score']['fullTime']['homeTeam']),
                     'away_team_score': int(match['score']['fullTime']['awayTeam']),
                     'away_team': match['awayTeam']['name']})
    # if no entry, skip
    except:
        pass 
df=pd.DataFrame(data=data)
df
```
